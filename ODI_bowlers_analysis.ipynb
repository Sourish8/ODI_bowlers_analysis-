<h3>Rename the column names appropriately</h3>

```python
df = df.rename(columns={"Span":"Active_Years",
                        "Mat":"Matches",
                        "Inns":"Innings",
                        "Balls":"Balls Bowled",
                        "Wkts":"Wickets",
                        "Ave":"Bowling average",
                        "Econ":"Economy",
                        "SR":"Strike_Rate",
                        "4": "4 Wickets",
                        "5" : "5 Wickets"})
display(df)
```

<h3> How many Players played for ICC?</h3>

```python
####splitting between name and country and placing them by creating new columns.

df[["Player_Name","Country"]] = df["Player"].str.split("(", expand =True)

#Replace the Closing Parenthesis ")" by empty string

df["Country"] = df["Country"].str.replace(")","")   

#Rearranging the columns

new_column = ["Player_Name","Country","Active_Years","Matches","Balls Bowled","Runs","Wickets","Bowling average","Economy","Strike_Rate","4 Wickets","5 Wickets"]
df= df[new_column]
```

```python
###Create a function for find the number of player played for ICC

def played_for_icc(x):
    if "ICC" in x:                                  
        return ("Yes")
    else:
        return("No")

#Create a new column based on how many players have played for the ICC and separate them with "yes" or "no"

df ["Played For ICC"] = df["Country"].apply(played_for_icc)
display(df.head())


#For Counting the values of ""Plyaed For ICC" column
df["Played For ICC"].value_counts() 
```

<p style="font-size: 20px;"> There are <b>13</b> players in this dataset who have played for ICC..</p>

<h3> How many different countries are present in this dataset?</h3>

```python
#replace the redundant data from the "Country" column with an empty string

df["Country"]=df["Country"].str.replace("ICC/","")
df["Country"]=df["Country"].str.replace("/ICC","")         
df["Country"]=df["Country"].str.replace("Asia/","")
df["Country"]=df["Country"].str.replace("Afr/","")

#find the unique value of the "country" column

different_countries = df["Country"].unique()

print(different_countries)
print("Number of different value = ",len(different_countries))
```

<p style="font-size: 20px;">There are 11 different Countries in this dataset.</p>

<h3>Which player(s) had played for the longest period of time?</h3>

```python
# Creating two new columns named "Start_year" and "End_year" by spliting 

df[["Start_Year","End_Year"]]= df["Active_Years"].str.split("-", expand=True)


# To change the data type of the newly created column

df[["Start_Year","End_Year"]] = df[["Start_Year","End_Year"]].astype("int64")

#Calculating the total period of playing time of a player by subtracting Start_Year from End_Year and insert them in new column.
df["Total_years_played"] = df["End_Year"] - df["Start_Year"] + 1


#Sorting "Total_years_played" column according to the decending order.

Sorted_Total_years_played = df.sort_values(by="Total_years_played", ascending = False)
display(Sorted_Total_years_played.head(10))
```

<p style="font-size: 20px;">The upper dataset shows the top players who had played for the longest period of time.</p>

<h3>Which player(s) had played for the shortest period of time?</h3>

```python
Shortest_period = df.sort_values(by="Total_years_played")
display(Shortest_period.head(10))
```

<p style="font-size: 20px;">The upper dataset shows the top players who had played for the shortest period of time.</p>
